```python
tick1_df = pd.read_csv('tickersdata/tickernum_0.csv', names=['index', 'data'])
```

```python
tick1_df = tick1_df.iloc[1:, :]
tick1_df = tick1_df['data']
```

```python
tick1_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
tick1_df[2]
```

```python
# creating a timestamps of 44 rows because is the highest number dividing 1364(shape of dataset)
x_train = []
y_train = []
for i in range(44,1364):
    x_train.append(tick1_df[i-44:i])
    y_train.append(tick1_df[i])
```

```python
x_train
```

```python
y_train
```

```python
X_train ,x_test, Y_train, y_test = train_test_split(x_train,y_train ,test_size=.2)
```

```python
X_train
```

```python
X_train,Y_train = np.array(X_train), np.array(Y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor='loss', patience=4)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(X_train,Y_train,epochs=100,batch_size=64, callbacks=[es])
```
