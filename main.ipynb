```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("training_data.csv")
data.tail()
```

```python
columns = data.columns
columns = columns[1:]
```

```python
for i in columns:
    ms = MinMaxScaler()
    conversion =np.array( data[i]).reshape(-1, 1)
    data[i] = ms.fit_transform(conversion)
```

```python
df = data.T
```

```python
df.columns = data.ticker.to_list()
```

```python
df = df.iloc[1:, :]
```

```python
df
```

```python
days_lst = df.index.to_list()
```

```python
df.at['day_0', 0]
```

# converting each ticker dataset to lists
 and saving them to different csv

```python
temp_lst = []
for i in range(0, 84):
    for j in days_lst:
        temp_lst.append(df.at[j,i])
        out_df = pd.DataFrame(temp_lst)
        out_df.to_csv('tickersdata/tickernum_' + str(i) + ".csv")
    temp_lst = []

print("Completed")
```

```python
tick1_df = pd.read_csv('tickersdata/tickernum_0.csv', names=['index', 'data'])
```

```python
tick1_df = tick1_df.iloc[1:, :]
tick1_df = tick1_df['data']
```

```python
tick1_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
tick1_df[2]
```

```python
# creating a timestamps of 44 rows because is the highest number dividing 1364(shape of dataset)
x_train = []
y_train = []
for i in range(44,1364):
    x_train.append(tick1_df[i-44:i])
    y_train.append(tick1_df[i])
```

```python
x_train
```

```python
y_train
```

```python
X_train ,x_test, Y_train, y_test = train_test_split(x_train,y_train ,test_size=.2)
```

```python
X_train
```

```python
X_train,Y_train = np.array(X_train), np.array(Y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor='loss', patience=5, verbose=True)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(X_train,Y_train,epochs=100,batch_size=64,validation_data=(np.array(x_test), np.array(y_test)))
```

```python
model.save('models/ticker1')
```

```python
preds = []
for i in range(0,(len(x_test)//43) ):
    temp_pred =model.predict(x_test[i])
    preds.append(temp_pred)
```
